*Para poder debuggear esta situación utilizaremos una tecnica muy tipica de pandas que se llama **boolean filtering**. Basicamente construimos una serie de booleanos con el mismo indice que nuestro dataframe(La serie clasicamente la llamaremos mask)*

```python
#Tomamos la columna problematica y vemos que peliculas superan el budget de el billon de dolares
movies_num['budget'] > 1e9
# Por lo que vemos son muy pocas
```

```python
# Averiguamos cuantas superan el billon contando los trues de la condicion
(movies_num['budget'] > 1e9).value_counts()
# y lo que nos arrojo es que solo existen 6 peliculas que superan el billon de dolares(mil millones)
# Lo que nos hace caer en cuenta que los datos que arroja el histograma no son correctos por estos datos que se salen del scope normal
# por lo tanto se debugean
```

```python
# Entonces lo que hacemos es tomar esta serie booleana y la guardamos en una variable para ver que peliculas son las que causan este problema
mask = (movies_num['budget'] > 1e9)
# y le decimos al dataset principal que no muestre estos datos
budget = movies[mask]
budget
```

## Combinamos nuestras tablas


```python
# Agregamos la columna title a nuestro dataframe movies_num, para utilizarla de llave para conectar el otro dataframe
#concatenamos
movies_num = pd.concat([movies_num,movies['movie_title']],axis=1)
movies_num.head()
#Podemos ver que ya se añadido la columna a nuestro dataframe
```

```python
#Haremos el merge de el dataframe movies_num con el de financials
movies_v2 = pd.merge(financials,movies_num,on='movie_title',how='left')
#Por lo que le hicimos un merge y se conserva la informacion de movies_num, y se pasa la de financials
movies_v2.shape
```

```python
movies_v2.head()
```

```python
movies_v2.columns
```

## Manejo de datos faltantes

```python
movies_v2 = movies_v2.drop_duplicates('movie_title') 
#REmovemos nombres duplicados
```

**Los datos faltantes generan problemas con muchos algoritmos de ML. Es por esto que existen distintas estrategias para lidiar con ellos.**

```python
## Para pedir ayuda con un comando podemos hacer algo asi
help(pd.Series.value_counts)
```

```python
movies_v2.notnull().apply(pd.Series.value_counts)
# Aplicamos es como un for que va columna por columna contando cuantos valores faltan
```

##### Podemos ver entonces que movie title no tiene ningun dato null mientras que Production Budget tiene 418 nulls

```python
(movies_v2 != 0).apply(pd.Series.value_counts)
```

#### Podemos ver cuantos valores son cero y cuantos no

No podremos entrenar nuestro con datos cuya variable 
objetivo no esta definida o sea nula(valor falso). Eliminemos esas líneas.

```python
available = ((movies_v2 != 0) & (movies_v2.notnull())) 
available.head()
```

```python
#Miramos linea por linea si es true o no
available.all(axis=1).value_counts()
```

```python
available.columns
```

```python
mask = available['Worldwide Gross']
```

```python
movies_v2 = movies_v2[mask]
```

```python
movies_v2['Worldwide Gross'].notnull().value_counts()
```

Tenemos una buena cantidad de datos de nuestra variable target, en el caso de las features que no son la variable objetivo una mejor solucion para lidiar con los datos faltantes es **reemplazar estos datos por otros que sean manejables** y no afecten la calidad de las predicciones. La estrategia más comun es utilizar la media de todos los ejemplos para la feature dada

```python
# Con el modulo preprocessing de sklearn para llenar nuestros datos faltantes en nuestro dataframe
from sklearn.preprocessing import Imputer
#Lo primero que hace es reconocer los datos faltantes, luego la estrategia con la que reemplazaremoslos campos nulos
imputer = Imputer(missing_values=np.nan, strategy='mean', axis=1)
```

```python
movies_v2
```

```python
#Primero remuevo columnas repetidas y columnas que no sean numericos
# y luego si relleno
movies_v2 = movies_v2.drop('movie_title',axis=1)
movies_v2.head()
```

```python
values = imputer.fit_transform(movies_v2)
values
```

```python
movies_v2.values
```

Podemos ver la diferencia entre el dataframe relleno y el dataframe que contiene datos null

```python
# Mi modelo ya limpio se llamara X
X = pd.DataFrame(values)
```

```python
# Doy nombre a mis columnas, cojiendo los nombres del dataset de movies_v2
X.columns = movies_v2.columns
```

```python
# Doy nombre a mis filas, cojiendo los nombres del dataset de movies_v2
X.index = movies_v2.index
```

```python
X.head()
```

```python
#Guardo mi dataframe para no tener  que generarlo cada vez que abra este archivo.
X.to_csv('X.csv',index=False)
```
